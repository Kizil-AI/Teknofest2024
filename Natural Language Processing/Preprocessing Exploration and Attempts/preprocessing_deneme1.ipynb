```python
import os
```

```python
DATA_PATH_JSON = "./NLP-Teknofest24/NLPJSON/"
DATA_PATH_TXT = "./NLP-Teknofest24/NLPTXT-1028 Etiketlenen Raporlar/"
```

```python
with open(DATA_PATH_TXT + "1/1 (1).txt") as f:
    print(f.read())
```

```python
import json
d_list = list()
for i in range(1,5):
    with open(DATA_PATH_JSON+str(i)+"/all.jsonl",'r', encoding="utf-8") as f:
        for a in f.readlines():
            d_list.append(json.loads(a))
```

```python
import pandas as pd
```

```python
meta_data = pd.DataFrame(d_list)
```

```python
meta_data = meta_data.drop("Comments",axis=1)
```

```python
meta_data = meta_data.sort_values("id")
```

```python
meta_data.shape
```

```python
meta_data["text"].iloc[0]
```

```python
import difflib

def string_benzerlik(str1, str2):
    # SequenceMatcher kullanarak iki string arasındaki benzerliği hesaplama
    benzerlik = difflib.SequenceMatcher(None, str1, str2).ratio()
    return benzerlik
```

```python
text_index_list = dict()
for folder in range(1,5):
    for _,_,dirs in os.walk(DATA_PATH_TXT+str(folder)+"/"):
        for file in dirs:
            with open(DATA_PATH_TXT + str(folder)+"/"+file) as f:
                t = f.read()
                for i in range(meta_data.shape[0]):
                    if(t == meta_data["text"].iloc[i]):
                        if(i in text_index_list): continue
                        text_index_list[i] = file
                        break

```

```python
meta_data = meta_data.reset_index().drop("index",axis=1)
meta_data["text_index"] = text_index_list
```

```python
nan_text_index_list = dict()
for folder in range(1,5):
    for _,_,dirs in os.walk(DATA_PATH_TXT+str(folder)+"/"):
        for file in dirs:
            with open(DATA_PATH_TXT + str(folder)+"/"+file) as f:
                print(file)
                t = f.read()
                for i in nan_rows:
                    sb = string_benzerlik(t,meta_data[meta_data["text_index"].isna()].loc[i]["text"])
                    if(sb>0.9):
                        if(i in nan_text_index_list): continue
                        nan_text_index_list[i] = [file,sb]
                        break
```
