```python
import fitz  # PyMuPDF

def extract_bold_words_from_pdf(pdf_path, start_page=12):
    # Open the PDF file
    doc = fitz.open(pdf_path)
    bold_words = []

    # Iterate through the pages starting from the specified page
    for page_num in range(start_page - 1, len(doc)):
        page = doc.load_page(page_num)
        blocks = page.get_text("dict")["blocks"]
        for block in blocks:
            if "lines" in block:
                for line in block["lines"]:
                    for span in line["spans"]:
                        if "bold" in span["font"].lower():  # Check if the text is bold
                            words = span["text"]
                            #print(words)
                            bold_words.append(words)
    
    return bold_words

# Path to the PDF file
pdf_path = './trk-tipdili-kilavuz.pdf'

def clean_words(words : list[str]) -> list[str]:
    new_words = []
    for w in words:
        if "-" in w: 
            continue
        w = w.replace(':','')
        if w[-1] == " ": w = w[:-1]
        if "," in  w: new_words.extend(w.split(", "))
        else: new_words.append(w)

    return new_words

# Extract bold words from the specified page onward
bold_words = extract_bold_words_from_pdf(pdf_path)

# Remove colons from the words
cleaned_bold_words = clean_words(bold_words)

# Print the cleaned bold words
print(cleaned_bold_words)

with open('cleaned_medical_words_list.txt', 'w', encoding='utf-8') as f:
    for word in cleaned_bold_words:
        f.write(word + '\n')
```

```python
import re
from itertools import product

def generate_variations(word):
    # Find all parts of the word that have optional characters within parentheses.
    parts = re.split(r'(\(.*?\))', word)
    
    # Create a list to store all the possible segments.
    segments = []
    
    for part in parts:
        if part.startswith('(') and part.endswith(')'):
            # If the part is within parentheses, it can either be included or excluded.
            optional = part[1:-1]  # Remove the parentheses.
            segments.append(['', optional])
        else:
            # Otherwise, just add the part as is.
            segments.append([part])
    
    # Generate all combinations of the segments.
    all_combinations = product(*segments)
    
    # Join the segments to form the complete words.
    variations = [''.join(combo) for combo in all_combinations]
    
    return variations

# Function to process a list of words with optional characters
def process_word_list(word_list):
    new_list = []
    for word in word_list:
        variations = generate_variations(word)
        new_list.extend(variations)
    return new_list


# Process the list
new_list = process_word_list(cleaned_bold_words)

with open('no_paranthesis_medical_words_list.txt', 'w', encoding='utf-8') as f:
    for word in new_list:
        f.write(word + '\n')
# Print the new list
print(new_list)
```
